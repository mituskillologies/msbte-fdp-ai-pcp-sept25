```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
df.shape
```

```python
x = df.iloc[:, [3, 4]]
x
```

```python
plt.figure(figsize=(16, 10))
plt.title("Customer Data")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.scatter(x.iloc[:,0], x.iloc[:, 1]);
```

```python
# kmeans
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters= 3, random_state= 0)
```

```python
labels = km.fit_predict(x)
```

```python
labels
```

```python
km.inertia_
```

#### Sum of squared error / Elbow Method

```python
inertia = []
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
for k in range(1, 31):
    km = KMeans(n_clusters= k, random_state= 0)
    km.fit_predict(x)
    inertia.append(km.inertia_)
```

```python
inertia
```

```python
# inertia vs number of clusters
```

```python
plt.figure(figsize=(16, 12))
plt.title('inertia vs k')
plt.xlabel('number of clusters')
plt.ylabel("inertia")
plt.xticks(range(0, 31, 1))
plt.plot(range(1, 31), inertia, marker = 'o', color = 'black', mfc = 'red');
plt.grid()
```

```python
# build the model with appropriate k value
```

```python
km = KMeans(n_clusters= 5, random_state= 0)
```

```python
labels = km.fit_predict(x)
```

```python
labels
```

```python
km.inertia_
```

```python
cluster_centers = km.cluster_centers_
cluster_centers;
```

```python
plt.figure(figsize=(16, 10))
plt.title("Customer Data")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.scatter(x.iloc[:,0], x.iloc[:, 1], c = labels);
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s = 200, 
            color = 'black');
```
