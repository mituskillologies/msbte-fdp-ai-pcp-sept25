```python
# Location: https://mitu.co.in/dataset
# File -> electronics.csv
```

```python
import pandas as pd
import csv
```

```python
dataset = []
with open('electronics.csv','r') as file:
    reader = csv.reader(file, delimiter = ',')
    for row in reader:
        dataset += [row]
```

```python
dataset
```

```python
pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
data = te.fit_transform(dataset)
```

```python
data
```

```python
te.columns_
```

```python
df = pd.DataFrame(data, columns= te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Step-1 Generate the frequent itemset
fre_itemset = apriori(df, min_support=0.25, use_colnames=True)
```

```python
fre_itemset
```

```python
rules = association_rules(fre_itemset, min_threshold=0.5, 
                          metric='confidence')
```

```python
rules
```

```python
rules = rules.iloc[:,[0,1,4,5]]
```

```python
rules
```

```python
rules.sort_values(by='confidence', ascending=False)
```

```python
rules[rules['antecedents'] == {'USB-C Cable'}]['consequents'].values[0]
```

```python
rules.to_csv('rules.csv')
```
