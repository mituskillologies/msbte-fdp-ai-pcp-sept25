```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.array([23, 34, 45, 50, 61])
y = np.array([12, 21, 29, 34, 42])
z = np.array([57, 49, 41, 32, 26])
```

```python
# line plot
```

```python
plt.title("line plot")
plt.xlabel('x values')
plt.ylabel('y values')
plt.plot(x, y, marker = 'o', mfc = 'black');
plt.grid()
# markers => o, >, <, ^, H, *
```

```python
# bar graph
```

```python
courses = ['python', 'kotlin', 'swift', 'go', 'dart']
students = [56, 61, 40, 55, 45]
```

```python
plt.title("student enrollment")
plt.xlabel('courses')
plt.ylabel('no. of students')
plt.bar(courses, students, color = 'black');
```

```python
# pie chart
```

```python
plt.pie(students, labels = courses, autopct = "%.2f");
```

```python
# scatter
```

```python
plt.scatter(x, y, color = 'red', marker ='*');
plt.grid()
```
